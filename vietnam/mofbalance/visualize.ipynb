```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import imageio
import os
```

```python
data = pd.DataFrame(pd.read_csv('https://raw.githubusercontent.com/thanhqtran/dataset/6fb7255cb089ba11f4f9f0b18f0d1d65ae6a49b6/vietnam/mofbalance/state_budget_balance.csv'))
# delete legends columns
df = data.drop(['Classification','item_VN','item'], axis=1)
# convert data to long form 
df
df_melted = df.melt(id_vars=['Code'], value_vars=[ year for year in df.columns[1:] ])
df_melted
```

```python
df
```

```python
# figure of agg_Exp and agg_Rev
inquiry = df_melted[df_melted['Code'] == 'agg_Exp']
inquiry2 = df_melted[df_melted['Code'] == 'agg_Rev']
inquiry3 = df_melted[df_melted['Code'] == 'def']
filenames = []
for i in range(len(inquiry['variable'])):
    plt.plot(inquiry['variable'][:i], inquiry['value'][:i].astype(float)/1000, label='Aggregate Expenditure')
    plt.plot(inquiry2['variable'][:i], inquiry2['value'][:i].astype(float)/1000, label='Aggregate Revenue')
    plt.plot(inquiry3['variable'][:i], inquiry3['value'][:i].astype(float)/1000, label='Deficit', color='red', linestyle=':')
    plt.ylabel('1000 Bil. VND')
    plt.xlabel('Year')
    plt.xticks(rotation=90)
    plt.legend()
    plt.title('Vietnamese Government Budget Balance')
    #create file name and add to list
    filename = f'{i}.png'
    filenames.append(filename)
    #save frames
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.close()
    # build gif

gif_filename = f'budget.gif'
images = []
duration = 0.15

with imageio.get_writer(gif_filename, mode='I', duration=duration, loop=1) as writer:
    for filename in filenames:
        image = imageio.imread(filename)
        writer.append_data(image)

# Remove files
for filename in set(filenames):
    os.remove(filename)    
```
